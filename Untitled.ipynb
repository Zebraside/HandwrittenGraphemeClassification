```python
class HandwrittenGraphemeDataset(Dataset):
    """Handwritten grapheme dataset."""

    def __init__(self, csv_file, parq_dir, transform=None):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.grapheme_frame = pd.read_csv(csv_file)
        self.parq_dir = parq_dir
        self.transform = transform

    def __len__(self):
        return len(self.landmarks_frame)

    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()

        img_name = self.grapheme_frame.iloc[idx].image_id
        image = io.imread(os.path.join(self.parq_dir,img_name))
        landmarks = self.landmarks_frame.iloc[idx, 1:]
        landmarks = np.array([landmarks])
        landmarks = landmarks.astype('float').reshape(-1, 2)
        sample = {'image': image, 'landmarks': landmarks}

        if self.transform:
            sample = self.transform(sample)

        return sample
```

```python
from fastai.vision import *
fold = 0
nfolds=4
label_cols = ['grapheme_root', 'vowel_diacritic', 'consonant_diacritic']
data = ImageList.from_df(
        train_df,
        path='',
        folder="Data",
        suffix='.png',
        cols='image_id',
        convert_mode='L') \
        .split_by_idx(range(fold * len(train_df) // nfolds, (fold + 1) * len(train_df) // nfolds)) \
        .label_from_df(cols=label_cols) \
        .transform(get_transforms(do_flip=False, max_warp=0.1), size=128, padding_mode='zeros') \
        .databunch(bs=128)
data = data.normalize((
              [0.0692],
              [0.2051],
          ))
```
